```python
# Importing packages
import pandas as pd
import glob
```

```python
# Reading and concatenating the source CSV files:

# Define the path to the CSV files
path = "D:/Github/Jupyter notebooks/World Happiness Score_Interactive Dashboard/Source_Datasets/*.csv"

# Use glob to get all the CSV files in the directory
all_files = glob.glob(path)

# Create an empty list to hold the DataFrames
df_list = []

# Loop through the list of files and read each one into a DataFrame
for file in all_files:
    df = pd.read_csv(file)
    # Extract the year from the filename
    year = int(file.split('_')[-1].split('.')[0])  # Adjust this based on your filename format
    df['year'] = year
    df_list.append(df)

# Concatenate all the DataFrames in the list into a single DataFrame
combined_df = pd.concat(df_list, ignore_index=True)

# Reorder the columns
combined_df = combined_df[['country', 'region', 'year', 'happiness_score', 'gdp_per_capita', 'social_support', 'healthy_life_expectancy', 'freedom_to_make_life_choices', 'generosity', 'perceptions_of_corruption']]
whr_df = combined_df.sort_values(['country','year'], ignore_index=True )
# Display the combined DataFrame
with pd.option_context('display.max_rows', None,
                       'display.max_columns', None
                       ):
    display(whr_df)

whr_df.to_csv('WHR_combined.csv')
```
