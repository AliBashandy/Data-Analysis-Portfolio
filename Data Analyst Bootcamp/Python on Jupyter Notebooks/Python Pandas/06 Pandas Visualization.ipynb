## Creating Visualizations using Pandas Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Github\Data-Analysis-Portfolio\Data Analyst Bootcamp\Python on Jupyter Notebooks\Python Pandas\Data\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
df.plot()
```

```python
df.plot( y = 'Flavor Rating')
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
df.plot(kind = 'line', subplots = True, title = ['As per Flavor', 'As per Texture', 'As Overall'])
```

```python
df.plot(kind = 'line', xlabel = 'Daily data', ylabel = 'Score')
```

```python
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'bar', stacked = True)
```

```python
df['Flavor Rating'].plot(kind = 'bar')
```

```python
df.plot.barh(stacked = True)
```

```python
df.plot.scatter(x = 'Flavor Rating', y = 'Overall Rating', s = 100, c = 'yellow')
```

```python
df.plot.hist(bins = 20)
```

```python
df.boxplot()
```

```python
df.plot.area()
```

```python
df.plot.area(figsize = (10,10))
```

```python
df.plot.pie(y = 'Overall Rating')
```
