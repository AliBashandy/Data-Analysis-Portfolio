## Group By and Aggregate Functions

```python
import pandas as pd
```

```python
df = pd.read_csv(r"D:\Github\Data-Analysis-Portfolio\Data Analyst Bootcamp\Python on Jupyter Notebooks\Python Pandas\Data\Flavors.csv")
df
```

```python
group_by_frame = df.groupby('Base Flavor')
```

```python
## Here will get the mean of the grouped only integers
group_by_frame.mean(['Flavor Rating',	'Texture Rating',	'Total Rating'])
```

```python
df.groupby('Base Flavor').mean(['Flavor Rating','Texture Rating','Total Rating'])
```

```python
## Here will get the count of grouped items
df.groupby('Base Flavor').count()
```

```python
## to get the max

df.groupby('Base Flavor').max()
```

```python
df.groupby('Base Flavor').min()
```

```python
df.groupby('Base Flavor').sum()
```

```python
## You can aggregate in multiple values
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean','max','min','count','sum']})
```

```python
df
```

```python
## Group by multiple 
df.groupby(['Base Flavor', 'Liked']).mean(['Flavor Rating','Texture Rating','Total Rating'])
```

```python
## Muliple agg on multi groupby
df.groupby(['Base Flavor', 'Liked']).agg({'Flavor Rating': ['mean','max','min','count','sum']})
```

```python
## .describe to get more details about the agg 
df.groupby('Base Flavor').describe()
```
