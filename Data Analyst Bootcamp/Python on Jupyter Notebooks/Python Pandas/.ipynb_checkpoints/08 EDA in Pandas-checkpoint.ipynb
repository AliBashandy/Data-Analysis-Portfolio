## Explore Data in Pandas 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Github\Data-Analysis-Portfolio\Data Analyst Bootcamp\Python on Jupyter Notebooks\Python Pandas\Data\world_population.csv")
```

```python
## this to change the float format
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
## Get the basic info of the data
df.info()
```

```python
## Get high level info about the data
df.describe()
```

```python
## The sum of the null values
df.isnull().sum()
```

```python
## To check the unique values
df.nunique()
```

```python
## Here we will sort by top countries in population basesd on 2022.
## By default its ascending order thats why we false it
## head() will show the top 5 by default
df.sort_values(by = "2022 Population", ascending = False).head(10)
```

```python
## Checking the correlation
df.corr(numeric_only = True)
```

```python
sns.heatmap(df.corr(numeric_only = True), annot = True)
plt.rcParams['figure.figsize'] = (20,5)
plt.show()
```

```python
df.groupby('Continent').mean(numeric_only = True).sort_values(by = "2022 Population", ascending = False)
```

```python
df.columns
```

```python
df[df['Continent'].str.contains('Oceania')]
```

```python
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean(numeric_only = True).sort_values(by = "2022 Population", ascending = False)
df2
```
