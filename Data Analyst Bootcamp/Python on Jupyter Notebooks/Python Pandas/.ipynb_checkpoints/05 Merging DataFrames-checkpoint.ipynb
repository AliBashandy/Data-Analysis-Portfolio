## Merge, Join, and Concatenate

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"D:\Github\Data-Analysis-Portfolio\Data Analyst Bootcamp\Python on Jupyter Notebooks\Python Pandas\Data\LOTR.csv")
```

```python
df1
```

```python
df2 = pd.read_csv(r"D:\Github\Data-Analysis-Portfolio\Data Analyst Bootcamp\Python on Jupyter Notebooks\Python Pandas\Data\LOTR 2.csv")
```

```python
df2
```

```python
## By default its "inner join" considering the dataframe called first is the left side
df1.merge(df2)
```

```python
## using "how" will allow to change the method of merging
df1.merge(df2, how = 'outer')
```

```python
## To specify the coulumn use "on"
df1.merge(df2, how = 'inner', on = 'FellowshipID')
```

```python
df1.merge(df2, how = 'inner', on = 'FellowshipID', suffixes = ('_left','_right'))
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
## "cross" will take every value from left data frame and concatenate it with left side
df1.merge(df2, how = 'cross')
```

```python
## In joins we have to define the suffix (left and Right)
df1.join(df2, on = 'FellowshipID', how = 'outer',lsuffix = '_left', rsuffix = '_right')
```

```python
## Joins works better with indexes

df3 = df1.set_index('FellowshipID')
df4 = df2.set_index('FellowshipID')
```

```python
df3.join(df4, how = 'outer',lsuffix = '_left', rsuffix = '_right')
```

```python
## Also we can set the index inside the join 

df5 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), how = 'outer',lsuffix = '_left', rsuffix = '_right')
df5
```

```python
## Concatenate like append in SQL put one data frame over the other

```

```python
## If we schoose the inner join it will take only the complete columns 
pd.concat([df1, df2], join = 'inner')
```

```python
## Defining the axis will put the data side by side depends on the indexes
pd.concat([df1, df2], join = 'outer', axis = 1)
```
